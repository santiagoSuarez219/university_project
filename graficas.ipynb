```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('medidas_ba√±o_liquido.csv')
df = df.astype(float)
```

```python
plt.figure(figsize=(6,6))
sns.boxplot(data=df,x="Patron", y="Medida")
plt.title('Medidas obtenidas ba√±o liquido')
plt.xlabel('Patron ¬∞C')
plt.ylabel('Medida ¬∞C')
plt.show()
```
